## données météorologiques avec une API

```python
liste_ville=[]
url = "https://nominatim.openstreetmap.org/search"

headers = {
    "User-Agent": "myapp" #peu importe le nom de user agent
}

for ville in villes:
    params = {
        "city": ville,       # Nom de la ville
        "country":"France", #Nom du pays
        "format": "json",  # Format de la réponse
        "limit": 1
    }
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    liste_ville.append(data)


liste_ville
```

```python

import pandas as pd
from tqdm import tqdm#affiche la barre de progreesion

data_ville=pd.DataFrame(columns=["idville","name","latitude","longitude"])
for i in tqdm(range(len(villes))):
  params = {
        "city": villes[i],       # Nom de la ville
        "country":"France", #Nom du pays
        "format": "json",  # Format de la réponse

    }
  response = requests.get(url, params=params, headers=headers)
  data_ville.loc[i,'idville']=i
  data_ville.loc[i,'name']=villes[i]
  data_ville.loc[i,'latitude']=response.json()[0]['lat']
  data_ville.loc[i,'longitude']=response.json()[0]['lon']

data_ville
```
